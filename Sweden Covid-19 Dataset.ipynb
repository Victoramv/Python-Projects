## Missing values

```python
pd.isnull(data['Region_Deaths'])
```

Show the not null values:

```python
pd.notnull(data['Region_Deaths'])
```

How many null values do we have?

```python
pd.isnull(data['Region_Deaths']).values.ravel().sum()
```

```python
data.dropna(axis=0, how='all')
```

```python
data["Region_Deaths"]=data["Region_Deaths"].fillna(0)
```

## Visualization

```python
%matplotlib inline
```
